```python
while 조건 : true 일 때 실행 됨
다시 한번 true면 실행 될 코드가 또 실행 됨
올라와서 조건으로 올라오게 되고 true인지 false인지 판별함. 어느 순간에 이 조건이 false가 나오게 되면 while문 탈출하고 while문 아래에 있는 코드드를 실행하게 됨

if 문과 비교하자면 while문은 true면 계속 반복한다.
```

```python
if 10 < 90:
  print("True")
# 10은 90보다 작은 결과는 True이기 때문에 결과인 print문이 출력 됨
```

```python
# while 10 < 90:
#   print("True")
#  10 이 90이라는 작다라는 명제는 변하지 않음. 즉 계속 True이므로 while문 내에서 무한 반복.
# 멈추려면 ctrl + c
```

```python
i = 0
while i < 6:
  print("True")
  i = i+1
# 
# i를 순차적으로 증가시켜주기
# True를 총 6번 반복시켜줌 - while loop 조건식에서 True 시 반복되고, False인 즉시 While loop 탈출
0에서 시작되었기 때문에 1씩 증가하면서 반복이 되게 되니까, 0, 1, 2, 3 ,4, 5..
6이 되는 순간 6이 6보다 작지 않기 때문에 false이므로 반복 종료
```

```python
i = 0
while i < 6:
  print(i)
  i = i+1
# i가 어떻게 증가 되는지 확인 할 수 있음. 
```

```python
# 내가 작성해보기
# 5부터 0까지 거꾸로 출력되는 프로그램 작성
i = 5
while i > 0: #  = while i > -1:
  print(i)
  i = i-1

```

```python
i = 0
while True :
  print(i)
  i = i + 1

  if i >= 3 :
    print("if문 동작")
    break

print("반복문 종료!")

# 0부터 하나씩 증가를 하다가 if문에 3보다 크거나 같다에 걸리게 되면서
# 동작을 하고 break를 통해 반복문 빠져나옴.
# break를 통해 내가 원하는 상황에 탈출 가능
```

```python
# continue 만나는 순간에 빠져나가는 것이 아닌 while 문 조건으로 올라가서 다시 한번 반복을 하게 됨
# continue 를 만나게 되면 다시 반복으로 올라가기 때문에 continue 
# 뒤에 있는 실행문은 실행이 안 됨.
```

```python
i = 0
while i < 10 :
  i += 1 # i는 하나씩 증가해나가겠지만,
  
  if i % 2 == 0: # i % 2 == 0인 경우에는 continue를 만나 해당문이 무시되고 그 위 조건문으로 다시 올라감
    continue
  print(i)
  
print("반복 종료!")
```

```python
# random : 임의의 난수를 만들어주는 기능
# sample : random의 범위와 갯수라는 데이터를 넣어주어야 함
# random범위는 1 - 100, 갯수 10이라고 하면, 1부터 100까지의 숫자들 중에
# 10개의 데이터를 랜덤으로 만들어준다라는 의미
```

```python
import random
randomNumber = random.sample(range(1, 101), 10) #random으로 sample하나 만들어 줘
# 1부터 100까지 숫자 중에 랜덤 숫자 10개를 뽑아줘
print(randomNumber)
```

```python
# 1 ~ 50 숫자 중에 임의로 3가지 샘플 뽑아줘
import random
randomNumber = random.sample(range(1, 51), 3) 
print(randomNumber)
```

```python
# 사용자에게 로또를 몇 개 살 건지 숫자 입력
# 1부터 45까지 숫자 중 6개를 랜덤으로 뽑기
# 사용자에게 입력받은 개수만큼 2를 뽑기
# 오름차순으로 정렬해서 출력하기
import random
lotto = int(input("로또를 몇 개 사실건가요??"))
for i in range(lotto) : #lotto 사는 갯수만큼 반복문
  randomNumber = random.sample(range(1,46), 6)
  randomNumber.sort()
  print(randomNumber)
```
